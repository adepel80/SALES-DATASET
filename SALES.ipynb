### REMOVE THE DOLLAR SIGN $ CALCULATE THE >,< OR = OF THE PRODUCTS

```python
sales['item_price'] = sales['item_price'].replace('''[$,]''', '', regex = True).astype(float)
```

```python
sales['item_price']
```

### CALCULATE PRODUCTS GREATER THAN 10 

```python
for x in sales['item_price']:
    if x > 10:
        print(x)
```

```python
sales
```

```python
len(sales[sales["item_price"] > 10])
```

```python
sales[sales["item_price"] > 10].count()
```

### CALCULATE PRODUCTS LESSER THAN OR EQUALS TO 10

```python
for x in sales['item_price']:
    if x <= 10:
        print(x)
```

```python
len(sales[sales["item_price"] <= 10])
```
